<p>možemo definirati svoju funkciju radi jednostavnijeg pisanja</p>

```python
def mjesoviti_produkt(v1,v2,v3):
    return v1.cross_product(v2).dot_product(v3)
```

<h2>Neki teoremi</h2>

<p>U ovom dijelu uvjerit ćemo se u istinitost nekih formula na način da vektore općenito zadamo njihovim koordinatama u ortonormiranoj bazi, a zatim raspišemo lijevu i desnu stranu jednakosti u tim koordinatama i provjerimo da su one zaista jednake.</p>

```python
var("a_x a_y a_z b_x b_y b_z c_x c_y c_z l")
```

```python
a=vector((a_x,a_y,a_z)); b=vector((b_x,b_y,b_z)); c=vector((c_x,c_y,c_z))
```

<p style="font-size:20px;">$(\vec{a}+\vec{b})^2=\vec{a}^2+2\vec{a}\vec{b}+\vec{b}^2$</p>

```python
a+b
```

```python
lijeva_strana=(a+b).dot_product(a+b); lijeva_strana
```

```python
desna_strana=a.dot_product(a)+2*a.dot_product(b)+b.dot_product(b); desna_strana
```

```python
expand(lijeva_strana-desna_strana)
```

<p style="font-size:20px;">$(\lambda\vec{a})\vec{b}=\lambda(\vec{a}\vec{b})$</p>

```python
lijeva_strana=(l*a).dot_product(b); lijeva_strana
```

```python
desna_strana=l*(a.dot_product(b)); desna_strana
```

```python
expand(lijeva_strana-desna_strana)
```

<p style="font-size:20px;">$\vec{a}\times(\vec{b}+\vec{c})=\vec{a}\times\vec{b}+\vec{a}\times\vec{c}$</p>

```python
lijeva_strana=a.cross_product(b+c); lijeva_strana
```

```python
desna_strana=a.cross_product(b)+a.cross_product(c); desna_strana
```

<p>dodatno sređivanje lijeve strane po komponentama</p>

```python
lijeva_strana.apply_map(lambda t: expand(t))
```

<p>razlike lijeve i desne strane je nulvektor, što znači da su vektori s lijeve i desne strane promatrane jednakosti jednaki</p>

```python
lijeva_strana.apply_map(lambda t: expand(t))-desna_strana
```

<p style="font-size:20px;">$(\vec{a}\times\vec{b})\times\vec{c}=(\vec{a}\vec{c})\vec{b}-(\vec{b}\vec{c})\vec{a}$</p>

```python
lijeva_strana=(a.cross_product(b)).cross_product(c)
lijeva_strana.apply_map(lambda t: expand(t))
```

```python
desna_strana=(a.dot_product(c))*b-(b.dot_product(c))*a
desna_strana.apply_map(lambda t: expand(t))
```

```python
lijeva_strana.apply_map(lambda t: expand(t))-desna_strana.apply_map(lambda t: expand(t))
```

<p style="font-size:20px;">$\vec{a}\times(\vec{b}\times\vec{c})=(\vec{a}\vec{c})\vec{b}-(\vec{a}\vec{b})\vec{c}$</p>

```python
lijeva_strana=a.cross_product(b.cross_product(c))
lijeva_strana.apply_map(lambda t: expand(t))
```

```python
desna_strana=(a.dot_product(c))*b-(a.dot_product(b))*c
desna_strana.apply_map(lambda t: expand(t))
```

```python
lijeva_strana.apply_map(lambda t: expand(t))-desna_strana.apply_map(lambda t: expand(t))
```

<p style="font-size:20px;">Jacobijev identitet: $(\vec{a}\times\vec{b})\times\vec{c}+(\vec{b}\times\vec{c})\times\vec{a}+(\vec{c}\times\vec{a})\times\vec{b}=\vec{0}$</p>

```python
lijeva_strana=(a.cross_product(b)).cross_product(c)+(b.cross_product(c)).cross_product(a)+(c.cross_product(a)).cross_product(b)
lijeva_strana.apply_map(lambda t: expand(t))
```

<p style="font-size:20px;">Lagrangeov identitet: $(\vec{a}\vec{b})^2+(\vec{a}\times\vec{b})^2=\vec{a}^2\cdot\vec{b}^2$</p>

```python
lijeva_strana=expand((a.dot_product(b))^2+(a.cross_product(b)).dot_product(a.cross_product(b))); lijeva_strana
```

```python
desna_strana=expand((a.dot_product(a))*(b.dot_product(b))); desna_strana
```

```python
lijeva_strana-desna_strana
```

<h2>Ortogonalna projekcija vektora na vektor</h2>

<p>Funkcija <span style="color: #ff0000;"><em>projekcija_vektor_vektor(x,a)</em></span> daje ortogonalnu projekciju vektora <span style="color: #ff0000;"><em>x</em></span> na vektor <span style="color: #ff0000;"><em>a</em></span>.</p>

```python
def projekcija_vektor_vektor(x,a):
    return (a.dot_product(x))/(norm(a)^2)*a
```

<div style="background-color: #F5FFFA; color:#000000; border: 2px solid #B0C4DE; border-radius: 15px; padding:0px 10px 25px; overflow:auto">
<h3>Zadatak</h3>
<p>Nađite ortogonalnu projekciju vektora $\vec{x}=(1,1,1)$ na vektor $\vec{a}=(2,1,2)$.</p>
</div>
<h3>Rješenje</h3>

```python
projekcija_vektor_vektor(vector((1,1,1)),vector((2,1,2)))
```

<h3>10. zadatak</h3>
<p>Zadani su vektori $\vec{a}=(2m,1,1-m)$, $\vec{b}=(-1,3,0)$ i $\vec{c}=(5,-1,8)$.</p>
<ul>
<li style="line-height:30px;">Odredite $m\in\mathbb{R}$ tako da vektor $\vec{a}$ zatvara jednake kutove s vektorima $\vec{b}$ i $\vec{c}$.</li>
<li style="line-height:30px;">Za tako pronađeni $m$ izračunajte volumen tetraedra određenog s vektorima $\vec{a},\vec{b},\vec{c}$ i duljinu visine tog tetraedra spuštenu na bazu određenu s vektorima $\vec{b}$ i $\vec{c}$.</li>
</ul>
</div>
<h3>Rješenje</h3>

```python
var("m")
a=vector((2*m,1,1-m))
b=vector((-1,3,0))
c=vector((5,-1,8))
```

```python
jednadzba=a.dot_product(b)/(sqrt(a.dot_product(a))*norm(b))==a.dot_product(c)/(sqrt(a.dot_product(a))*norm(c))
jednadzba
```

```python
solve(jednadzba,m)
```

<p>volumen tetraedra</p>

```python
V=1/6*abs(mjesoviti_produkt(a.subs(m=1/4),b,c)); V
```

<p>površina baze određena s vektorima $\vec{b}$ i $\vec{c}$</p>

```python
B=1/2*norm(b.cross_product(c)); B
```

<p>duljina visine tetraedra spuštena na bazu određenu vektorima $\vec{b}$ i $\vec{c}$</p>

```python
3*V/B
```

<ul>
<li style="line-height:30px;">Nađite parametar $k$ ako znamo da su vektori $\vec{b}$ i $\vec{c}$ okomiti.</li>
<li style="line-height:30px;">Odredite površinu paralelograma određenog s vektorima $\vec{a}$ i $\vec{c}$.</li>
<li style="line-height:30px;">Odredite volumen piramide određene s vektorima $\vec{a}$, $\vec{b}$ i $\vec{c}$.</li>
<li style="line-height:30px;">Da li vektori $\vec{a},\vec{b},\vec{c}$ čine bazu za $V^3$? Ako je odgovor potvrdan, prikažite vektor $\vec{d}=(-1,4,6)$ kao linearnu kombinaciju vektora $\vec{a}$, $\vec{b}$ i $\vec{c}$.</li>
</ul>
</div>
<h3>Rješenje</h3>

```python
var("k")
a=vector((3,-1,4))
b=vector((-1,2,5))
c=vector((2,-2,k))
d=vector((-1,4,6))
```

```python
solve(b.dot_product(c)==0,k)
```

<p>površina paralelograma</p>

```python
norm(a.cross_product(c.subs(k=6/5)))
```

<p>volumen piramide</p>

```python
1/6*abs(mjesoviti_produkt(a,b,c.subs(k=6/5)))
```

<p style="line-height:30px;">Kako je mješoviti produkt vektora $\vec{a}, \vec{b}, \vec{c}$ različit od nule, slijedi da su oni nekomplanarni pa čine bazu za $V^3$. Stoga se vektor $\vec{d}$ može na jedinstveni način prikazati kao linearna kombinacija vektora $\vec{a}, \vec{b}, \vec{c}$.</p>
<p>$\vec{d}=x\cdot\vec{a}+y\cdot\vec{b}+z\cdot\vec{c}$</p>

```python
d1=x*a+y*b+z*c.subs(k=6/5); d1
```

```python
solve([d1[0]==d[0],d1[1]==d[1],d1[2]==d[2]],[x,y,z])
```
