<h3>22. zadatak</h3>
<p>Riješite jednadžbu $x^8+3x^4+2x^2+3=0$.</p>

<h3>Rješenje</h3>
<p style="line-height:25px;">SAGE nema automatiziran postupak korjenovanja kompleksnih brojeva, što se dolje vidi iz danih rješenja na izlazu.</p>

```python
var('x')
jednadzba=x^8+3*x^4+2*x^2+3==0
solve(jednadzba,x)
```

<p style="line-height:25px;">Stoga ćemo uvesti najprije supstituciju $x^2=t$ i riješiti dobivenu jednadžbu za $t$. Nakon toga ćemo pomoću naše funkcije <span style="color: #ff0000;"><em>kompleksni_korijen</em></span> izvaditi kvadratne korijene iz dobivenih $t$-ova.</p>

```python
var('t')
nova=jednadzba.substitute(x=sqrt(t)); nova
```

```python
rj=solve(nova,t); rj
```

```python
rj=list(map(lambda y:y.rhs(),rj)); rj
```

<p style="line-height:30px;">Primijenimo sada <span style="color: #ff0000;"><em>kompleksni_korijen</em></span> na dobivene $t$-ove kako bismo dobili rješenja za $x$-ove. Naime, $x=\sqrt{t}$. Radi preglednosti primjenjivat ćemo funkciju <span style="color: #ff0000;"><em>kompleksni_korijen</em></span> korak po korak, zasebno za svaki dobiveni $t$.</p>

```python
kompleksni_korijen(rj[0],2,sredi=True)
```

```python
kompleksni_korijen(rj[1],2,sredi=True)
```

```python
kompleksni_korijen(rj[2],2,sredi=True)
```

```python
kompleksni_korijen(rj[3],2,sredi=True)
```

Usput možemo primijeniti i funkciju <span style="color: #ff0000;"><em>flatten</em></span> kako bismo izbacili sve unutarnje zagrade (koje stvara funkcija <span style="color: #ff0000;"><em>kompleksni_korijen</em></span>) u listi rješenja za $x$-ove.</p>

<h3>6. primjer</h3>
<p>Pomoću Cardanove formule riješite jednadžbu $x^3-2x^2-x+2=0$.</p>
</div>
<p>Ova jednadžba ima sva rješenja cjelobrojna.</p>

```python
var('x y')
solve(x^3-2*x^2-x+2==0,x)
```

<p style="line-height:30px;">Međutim, Cardanova formula će dati vrlo komplicirane zapise tih rješenja. Da bismo mogli primijeniti Cardanovu formulu, moramo se prvo riješiti kvadratnog člana. U jednadžbi $x^3+ax^2+bx+c=0$ kvadratnog člana se rješavamo pomoću supstitucije $x=y-\frac{a}{3}$.</p>

```python
jed=x^3-2*x^2-x+2==0
jed.substitute(x=y+2/3)
```

```python
expand(jed.substitute(x=y+2/3))
```

<p style="line-height:30px;">Sada rješavamo jednadžbu $y^3-\frac{7}{3}y+\frac{20}{27}=0$ pomoću Cardanove formule.</p>

```python
rj=Cardan(-7/3,20/27); rj
```

<p>Sva rješenja početne jednadžbe su</p>

```python
rj2=list(map(lambda h:h+2/3,rj)); rj2
```

<p style="line-height:25px;">Dakle, užasno komplicirani zapisi rješenja početne jednadžbe, a znamo da su ta rješenja jako lijepa i cjelobrojna kao što smo i na početku vidjeli. No, možemo se numerički uvjeriti da su ovi komplicirani zapisi zaista jednaki $2,\, 1$ i $-1$.</p>

```python
list(map(lambda h:n(h,digits=20),rj2))
```

```python
list(map(lambda h:n(h,digits=80),rj2))
```
